```python
df = pd.read_csv('data.csv')
```

```python
Physics_Grade = list(df['Math Grade'].sample(frac=1))
Physics_Grade = pd.DataFrame(Physics_Grade, columns=['Physics Grade'])

Biology_Grade = list(df['Math Grade'].sample(frac=1))
Biology_Grade = pd.DataFrame(Biology_Grade, columns=['Biology Grade'])

History_Grade = list(df['Math Grade'].sample(frac=1))
History_Grade = pd.DataFrame(History_Grade, columns=['History Grade'])

CS_Grade = list(df['Math Grade'].sample(frac=1))
CS_Grade = pd.DataFrame(CS_Grade, columns=['CS Grade'])
```

```python
New_GPA = list(df.GPA.sample(frac=1))
New_GPA = pd.DataFrame(New_GPA, columns=['New_GPA'])
df = pd.concat([df,New_GPA],axis=1)
```

```python
df = pd.concat([df,Physics_Grade,Biology_Grade,History_Grade,CS_Grade],axis=1)
```

```python
df.drop_duplicates(inplace=True)
```
